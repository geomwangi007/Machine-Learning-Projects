```python
 import pandas as pd
 import numpy as np
 from sklearn.model_selection import train_test_split  , cross_val_score
 from sklearn.linear_model import LogisticRegression
 from sklearn.svm import SVC
 from sklearn.metrics import accuracy_score, confusion_matrix ,recall_score ,precision_score, f1_score
```

```python
#Separating the target and feature variables
x = sonar.drop(columns ='R' , axis = 1)

y = sonar['R']

```

```python
x
```

```python
y
```

```python
x_train, x_test ,y_train ,y_test = train_test_split( x ,y , test_size= 0.1 ,random_state= 42 , stratify= y)
```

```python
print(x.shape , x_train.shape , x_test.shape)
```

```python
models = [LogisticRegression() , SVC(kernel= 'linear')]
```

```python
def compare_models():

  for model in models:

    model.fit(x_train , y_train)

    x_train_pred = model.predict(x_train)

    x_test_pred = model.predict(x_test)

    x_train_accuracy = accuracy_score(x_train_pred , y_train)

    x_test_accuracy = accuracy_score(x_test_pred , y_test)


    print('Accuracy of the training data in a ', model ,' = ' ,np.round(x_train_accuracy*100,2) )

    print('Accuracy of the testing data in a ' , model, '= ' ,np.round(x_test_accuracy*100,2) )

    print('-----------------------------------------------------------------------------------')
```

```python
compare_models()
```

```python
def cross_val_comparison():

  for model in models:

    cv_score = cross_val_score(model , x  , y  , cv = 5)

    mean_accuracy = np.round(np.mean(cv_score)*100, 2)

    print('The Cross validation accuracies of ' , model , 'are' , ' = ' , cv_score)

    print('The mean accuracy of ' , model , 'is ' , mean_accuracy)

    print('-------------------------------------------------------------------------')


```

```python
cross_val_comparison()
```
