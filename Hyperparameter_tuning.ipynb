GridSearchCV & RandomizedSearchCV

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
from sklearn.datasets import load_breast_cancer

cancer_data = load_breast_cancer()

df = pd.DataFrame(cancer_data.data, columns = cancer_data.feature_names)
```

```python
df["Label"] = cancer_data.target
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df["Label"].value_counts()
```

0----m
1----b

```python
x = df.drop(columns = "Label", axis = 1)
y = df["Label"]
```

```python
x
```

```python
y
```

```python
x= np.asarray(x)
y = np.asarray(y)
```

#**GridSearchCV**


```python
model = SVC()
```

```python
#Hyperparameters

parameters = {"kernel" : ["linear","rbf","sigmoid"], "C" : [1, 5 , 10, 20]}
```

```python
#GridSearch
classifier = GridSearchCV(model , parameters , cv = 5)
```

```python
#Data fitting

classifier.fit(x,y)
```

```python
classifier.cv_results_
```

```python
best_parameters = classifier.best_params_
print(best_parameters)
```

```python
#Highest accuracy for the best parameters

Highest_accuracy = classifier.best_score_
Highest_accuracy
```

```python
#Loading the results as a  pandas df

result = pd.DataFrame(classifier.cv_results_)
result.head()
```

```python
grid_search_result = result[["param_C","param_kernel","mean_test_score"]]
grid_search_result
```

Highest Accuracy = 95.2%

Best Parameters = {'C':10, 'kernel':'linear'}

#**RandomizedSearchCV**

```python
model = SVC()
```

```python
#Hyperparameters

parameters = {"kernel" : ["linear","rbf","sigmoid"], "C" : [1, 5 , 10, 20]}
```

```python
#RandomizedSearch
classifier = RandomizedSearchCV(model , parameters , cv = 5)
```

```python
classifier.fit(x,y)
```

```python
classifier.cv_results_
```

```python
best_parameters = classifier.best_params_
print(best_parameters)
```

```python
#Highest accuracy for the best parameters

Highest_accuracy = classifier.best_score_
Highest_accuracy
```

```python
#Loading the results as a  pandas df

result = pd.DataFrame(classifier.cv_results_)
result.head()
```

```python
randomized_search_result = result[["param_C","param_kernel","mean_test_score"]]
randomized_search_result
```

Highest Accuracy = 95.2%

Best Parameters = {'C':10, 'kernel':'linear'}
