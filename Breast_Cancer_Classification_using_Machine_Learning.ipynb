Machine learning clasification project of cancer patients.I will be using a Fine needle aspiration procedure, a biopsy procedure.


```python
#Importing the dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
#Data Imporatation
data = pd.read_csv("/content/cancer_data.csv")
data.head()
```

```python
data.shape
```

```python
#Checking for null values
data.isnull().sum()
```

```python
cleaned_data = data.dropna(axis = 1)
cleaned_data.head()
```

```python
cleaned_data = cleaned_data.drop(columns = "id")
cleaned_data
```

```python
replace_diagnosis = {"B":0, "M":1}


# Replace values in column diagnosis using replace()
cleaned_data.replace(replace_diagnosis, inplace=True)
cleaned_data.head()
```

```python
cleaned_data["diagnosis"].value_counts()
```

```python
cleaned_data.shape
```

```python
cleaned_data["diagnosis"].value_counts()
```

B(Benign tumor)------->Non-cancerous
M(Malignant tumor)----->Cancerous

```python
#Separating the features and target

target = cleaned_data["diagnosis"]
features = cleaned_data.drop(columns="diagnosis")
```

```python
target
```

```python
features
```

```python
x_train,x_test,y_train,y_test = train_test_split(features ,target ,test_size = 0.2 ,random_state= 2)
```

```python
print(features.shape ,x_train.shape,x_test.shape)
```

```python
model = LogisticRegression()
```

```python
#Fitting the model to the training data
model.fit(x_train,y_train)
```

```python
#Model prediction

x_train_prediction = model.predict(x_train)

#Calculating the training data accuracy
training_data_accuracy = accuracy_score(y_train,x_train_prediction)
print("Training data accuracy score: ",training_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test)

#Calculating the testing data accuracy
test_data_accuracy = accuracy_score(y_test,x_test_prediction)
print("Testing data accuracy score: ",test_data_accuracy)
```
